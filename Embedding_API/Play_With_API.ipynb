```python
import os
from google import genai
from dotenv import load_dotenv
load_dotenv()

client = genai.Client(api_key=os.getenv("GEMINI_APIKEY"))
```

```python
generated_responce = client.models.generate_content(
    model="gemini-2.5-flash-lite-preview-09-2025",
    contents="Hello",
)
generated_text = generated_responce.text

print(generated_text)
```

```python
embedded_responce = client.models.embed_content(
    model="gemini-embedding-001",
    contents=["Hello, World! My name is SUPERUSER!!!", "HEEEEEELLOOOOO"],
)
embedded_vectors = embedded_responce.embeddings
print(len(embedded_vectors))
```

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter, Language

text_splitter = RecursiveCharacterTextSplitter(
    chunk_size=1500*4,
    chunk_overlap=200*4,
    separators=["\n\n", "\n", " ", ""],
    keep_separator=False,

)
```

```python
import fitz

with fitz.open("file.pdf") as pdf:
    text = ""
    for page in pdf:
        content = page.get_text()
        text += content
        print(text)

len(text_splitter.split_text(text))
text_splitter.split_documents
```

```python
from docx import Document
doc = Document("file.docx")
text = "\n".join([para.text for para in doc.paragraphs])
print(text_splitter.split_text(text)[1])
```
